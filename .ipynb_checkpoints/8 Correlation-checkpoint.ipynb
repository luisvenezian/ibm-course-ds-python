```python
import pandas as pd
df = pd.read_csv('./datasets/df_imports-85.csv')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['engine-size'] = pd.to_numeric(df['engine-size'], errors='coerce')
df[['price','engine-size']].dtypes
```

## Positive Linear Correlation

```python
import seaborn as sns
from matplotlib import pyplot as plt
eng_x_price = sns.regplot(x = 'engine-size', y = 'price', data = df)
plt.ylim(0, )
```

## Negative Linear Correlation

```python
mpg_x_price = sns.regplot(x = 'highway-mpg', y = 'price', data = df)
plt.ylim(0, )
```

### Not good to predict atribute

```python
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'], errors='coerce')
peak_x_price = sns.regplot(x = 'peak-rpm', y = 'price', data = df)
plt.ylim(0, )
```
